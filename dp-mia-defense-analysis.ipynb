```python
from itertools import groupby
from dataclass_csv import DataclassReader
from dataclasses import dataclass
from dataclasses import field
from operator import attrgetter

import matplotlib.pyplot as plt
import numpy as np
```

```python
@dataclass
class Experiment:
    epsilon: float
    delta: float
    batch_size: int
    max_grad_norm: float
    cifar_label: int
    victim_accuracy: float
    attack_accuracy: float
```

```python
with open('dp-grid.csv', mode='r') as csv_file:
    reader = DataclassReader(csv_file, Experiment)
    results = list(reader)
```

```python
x = [r.attack_accuracy for r in results]
y = [r.victim_accuracy for r in results]
colors = [r.epsilon for r in results]

# Create scatter plot
plt.scatter(x, y, c=colors, alpha=0.7, cmap='viridis')
plt.xlabel('Attack model accuracy')
plt.xlim(45, 60)
plt.ylabel('Victim model accuracy')
plt.ylim(0, 100)
plt.title('Impact of epsilon on victim and attack models')
plt.colorbar(label='Epsilon values')
plt.show()
```

```python
x = [r.attack_accuracy for r in results]
y = [r.victim_accuracy for r in results]
colors = [r.batch_size for r in results]

# Create scatter plot
plt.scatter(x, y, c=colors, alpha=0.7, cmap='viridis')
plt.xlabel('Attack model accuracy')
plt.xlim(45, 60)
plt.ylabel('Victim model accuracy')
plt.ylim(0, 100)
plt.title('Impact of batch size on victim and attack models')
plt.colorbar(label='Batch sizes')
plt.show()
```
